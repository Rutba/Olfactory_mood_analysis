```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
def modify_stimuli(molecules, behavior_descr):
    print("Modifying/cleaning stimuli …")
    # rename index to a common key
    molecules = molecules.reset_index().rename(columns={'index':'molecule_id'})
    behavior_descr = behavior_descr.reset_index().rename(columns={'index':'molecule_id'})
    # merge
    stim = molecules.merge(behavior_descr, on='molecule_id', how='left')
    
    # --- Fix: Handle missing or differently named SMILES column ---
    smiles_candidates = [c for c in stim.columns if 'IsomericSMILES' ]
    if len(smiles_candidates) > 0:
        smiles_col = smiles_candidates[0]
        print(f"Detected SMILES column: '{smiles_col}'")
        stim = stim.dropna(subset=[smiles_col])
    else:
        print("No SMILES column found — skipping SMILES filtering.")
        smiles_col = None
    
    # Drop duplicates if any
    stim = stim.drop_duplicates(subset=['molecule_id'])
   
    print(f"Stimuli after merge: {stim.shape}")
    stim.to_csv('stimuli_modified.csv', index=False)
    return stim
'''
def modify_stimuli(molecules, behavior_descr):
    print("Modifying/cleaning stimuli …")
    # rename index to a common key
    molecules = molecules.reset_index().rename(columns={'index':'molecule_id'})
    behavior_descr = behavior_descr.reset_index().rename(columns={'index':'molecule_id'})
    # merge
    stim = molecules.merge(behavior_descr, on='molecule_id', how='left')
    # drop duplicates/missing
    stim = stim.dropna(subset=['IsomericSMILES'])  # drop rows missing SMILES
    print(f"Stimuli after merge: {stim.shape}")
    stim.to_csv('stimuli_modified.csv', index=False)
    return stim
'''
```

```python
def modify_behavior(behav_df):
    print("Modifying/cleaning behaviour data …")
    # We assume long format: each row = participant × odor
    # Suppose columns include 'molcode' or 'odor_id' referring to odor stimulus
    # rename to 'molecule_id' to match stimuli
    if 'molcode' in behav_df.columns:
        behav_df = behav_df.rename(columns={'molcode':'molecule_id'})
    if 'odor_id' in behav_df.columns and 'molecule_id' not in behav_df.columns:
        behav_df = behav_df.rename(columns={'odor_id':'molecule_id'})
   
    # Identify rating columns (if 'pleasantness' not present, pick first numeric column)
    rating_col = 'pleasant' if 'pleasant' in behav_df.columns else None
    if rating_col is None:
        numeric_cols = behav_df.select_dtypes(include=np.number).columns.tolist()
        if numeric_cols:
            rating_col = numeric_cols[0]  # pick the first numeric column
            print(f"Warning: 'pleasantness' not found. Using '{rating_col}' instead as key rating.")
        else:
            raise KeyError("No numeric columns found in behavior data to use as ratings.")
    
    
    # drop rows missing key ratings
    #behav_df = behav_df.dropna(subset=['pleasantness'])
    behav_df = behav_df.dropna(subset=[rating_col])
    behav_df.to_csv('behavior_modified.csv', index=False)
    print(f"Behaviour rows: {behav_df.shape}")
    return behav_df
```

```python
'''
def merge_behavior_stimuli(behav_df, stim_df):
    print("Merging behaviour + stimuli …")
    merged = behav_df.merge(stim_df, how='left', on='molecule_id')
    merged = merged.dropna(subset=['smiles'])  # ensure matched
    print(f"Merged dataset: {merged.shape}")
    merged.to_csv('merged_data.csv', index=False)
    return merged
'''
def merge_behavior_stimuli(behav_df, stim_df):
    print("Merging behaviour + stimuli …")

    # --- Ensure both molecule_id columns exist ---
    if 'molecule_id' not in behav_df.columns:
        # Try to guess the key column in behavior data
        possible_keys = [c for c in behav_df.columns if 'odor' in c.lower() or 'molecule' in c.lower()]
        if possible_keys:
            behav_df = behav_df.rename(columns={possible_keys[0]: 'molecule_id'})
            print(f"Renamed {possible_keys[0]} → molecule_id in behavior data")
        else:
            raise KeyError("No molecule identifier column found in behavior data!")

    if 'molecule_id' not in stim_df.columns:
        stim_df = stim_df.reset_index().rename(columns={'index': 'molecule_id'})
        print("Added molecule_id to stimuli data from index")

    # --- Fix dtype mismatch ---
    behav_df['molecule_id'] = behav_df['molecule_id'].astype(str)
    stim_df['molecule_id'] = stim_df['molecule_id'].astype(str)

    # --- Merge datasets ---
    merged = behav_df.merge(stim_df, how='left', on='molecule_id')

    # --- Drop rows without valid structure if present ---
    smiles_cols = [c for c in merged.columns if 'IsomericSMILES']
    if smiles_cols:
        merged = merged.dropna(subset=[smiles_cols[0]])
        print(f"Dropped rows missing {smiles_cols[0]}")

    print(f"Merged dataset: {merged.shape}")
    merged.to_csv('merged_data.csv', index=False)
    return merged

```

```python
def preprocess_merged(df):
    print("Preprocessing merged data …")
    scaler = StandardScaler()
    # Example: scale pleasantness, intensity, familiarity if present
    for col in ['pleasant', 'intensive', 'familiar']:
        if col in df.columns:
            df[col + '_z'] = scaler.fit_transform(df[[col]])
    # Stimulus features: e.g., descriptor columns in stim_df that start with some prefix
    descriptor_cols = [c for c in df.columns if c.startswith('descriptor_')]
    if descriptor_cols:
        df[descriptor_cols] = scaler.fit_transform(df[descriptor_cols])
        # PCA for dimensionality reduction of descriptors
        pca = PCA(n_components=2, random_state=0)
        pcs = pca.fit_transform(df[descriptor_cols])
        df['pc1'] = pcs[:,0]
        df['pc2'] = pcs[:,1]
        print("Explained variance ratio (descriptor PCA):", pca.explained_variance_ratio_)
    df.to_csv('merged_preprocessed.csv', index=False) 
    return df
```

```python
def exploratory_analysis(df):
    print("Doing EDA …")
    if 'pleasant_z' in df.columns:
        sns.histplot(df['pleasant_z'], kde=True)
        plt.title("Distribution of Pleasant (z-score)")
        plt.show()
    if 'intensive_z' in df.columns:
        sns.scatterplot(data=df, x='intensive_z', y='pleasant_z', hue='sex' if 'sex' in df.columns else None)
        plt.title("Pleasantness vs Intensity by gender")
        plt.show()
    # Correlation
    cols = [c for c in df.columns if c.endswith('_z')]
    print("Correlations of z-scored variables:")
    print(df[cols].corr())
```

```python
def run_regression(df):
    print("Running regression …")
    # Example model: pleasantness_z ~ intensity_z + familiarity_z + pc1 + C(gender) + age
    formula = "pleasant_z ~ intensive_z + familiar_z"
    if 'pc1' in df.columns:
        formula += " + pc1"
    if 'sex' in df.columns:
        formula += " + C(sex)"
    if 'age' in df.columns:
        formula += " + age"
    model = smf.ols(formula, data=df).fit()
    print(model.summary())
    with open('model_summary.txt','w') as f:
        f.write(model.summary().as_text())
    return model
```

```python
def main():
    # Step 1: Stimuli
    molecules = pd.read_csv("molecules.csv") 
    behavior_descr = pd.read_csv("behavior_descr.csv")
    stim_df = modify_stimuli(molecules, behavior_descr)

    # Step 2: Behaviour
    behav_df = pd.read_csv("behavior_data.csv", low_memory=False)
    behav_mod = modify_behavior(behav_df)

    # Step 3: Merge
    merged_df = merge_behavior_stimuli(behav_mod, stim_df)

    # Step 4: Preprocess
    preproc_df = preprocess_merged(merged_df)
    #print(preproc_df.columns)
    # Step 5: EDA
    exploratory_analysis(preproc_df)

    # Step 6: Regression
    _ = run_regression(preproc_df)

    print("Project complete.")

if __name__ == "__main__":
    main()
```
